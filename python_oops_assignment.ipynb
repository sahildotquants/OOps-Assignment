## Theory Questions


### 1. What is Object-Oriented Programming (OOP)?

OOP is a programming paradigm based on the concept of **objects**, which bundle together data (attributes) and behavior (methods). It relies on four core pillars: encapsulation, abstraction, inheritance, and polymorphism.

### 2. What is a class in OOP?

A class is a **blueprint/template** for creating objects. It defines the attributes and methods that its objects will have.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand
```

### 3. What is an object in OOP?

An object is an **instance of a class** — a concrete entity created from the class blueprint, with its own data.

```python
my_car = Car("Toyota")   # my_car is an object of class Car
```

### 4. What is the difference between abstraction and encapsulation?

- **Abstraction** hides *implementation complexity* and shows only relevant functionality to the user (what an object does).
- **Encapsulation** hides the *internal data* of an object by restricting direct access (bundling data + methods, using private/protected attributes).

Example: A car's steering wheel is abstraction (you just drive); the engine internals being hidden inside the hood is encapsulation.

### 5. What are dunder methods in Python?

Dunder ("double underscore") methods are special built-in methods with names like `__init__`, `__str__`, `__len__`, `__add__`. They let objects integrate with Python's built-in syntax/operators.

```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y
    def __str__(self):
        return f"({self.x}, {self.y})"

print(Point(1, 2))   # (1, 2)
```

### 6. Explain the concept of inheritance in OOP.

Inheritance allows a class (child/subclass) to acquire attributes and methods from another class (parent/superclass), promoting code reuse.

```python
class Animal:
    def speak(self):
        print("Some sound")

class Dog(Animal):
    pass

Dog().speak()   # inherited from Animal
```

### 7. What is polymorphism in OOP?

Polymorphism means "many forms" — the same method name can behave differently depending on the object calling it.

```python
class Cat:
    def sound(self):
        return "Meow"

class Dog:
    def sound(self):
        return "Bark"

for animal in [Cat(), Dog()]:
    print(animal.sound())
```

### 8. How is encapsulation achieved in Python?

By prefixing attributes with an underscore (`_protected`) or double underscore (`__private`, triggers name-mangling), and exposing controlled access via methods or `@property`.

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance   # private

    def get_balance(self):
        return self.__balance
```

### 9. What is a constructor in Python?

The constructor is the `__init__` method, automatically called when an object is created, used to initialize its attributes.

```python
class Person:
    def __init__(self, name):
        self.name = name

p = Person("Sahil")   # __init__ runs automatically
```

### 10. What are class and static methods in Python?

- A **class method** (`@classmethod`) receives the class itself (`cls`) as the first argument and can access/modify class-level state.
- A **static method** (`@staticmethod`) receives no implicit first argument — it behaves like a plain function grouped inside the class.

```python
class Demo:
    @classmethod
    def cls_method(cls):
        return f"Called on {cls}"

    @staticmethod
    def static_method():
        return "No access to class or instance"
```

### 11. What is method overloading in Python?

Python doesn't support traditional method overloading (same name, different signatures) like Java/C++. Instead, it's simulated using default arguments or `*args`.

```python
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

Calculator().add(1, 2)       # works
Calculator().add(1, 2, 3)    # also works
```

### 12. What is method overriding in OOP?

When a child class provides its own implementation of a method already defined in its parent class.

```python
class Animal:
    def speak(self):
        print("Generic sound")

class Dog(Animal):
    def speak(self):        # overrides parent method
        print("Bark!")
```

### 13. What is a property decorator in Python?

`@property` lets a method be accessed like an attribute (no parentheses), commonly used to add getter/setter logic while keeping a clean interface.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def area(self):
        return 3.14159 * self._radius ** 2

c = Circle(5)
print(c.area)   # accessed like an attribute, not c.area()
```

### 14. Why is polymorphism important in OOP?

It allows a single interface to work with different underlying object types, making code more flexible, extensible, and easier to maintain (e.g., writing generic code that works for many subclasses without knowing their exact type).

### 15. What is an abstract class in Python?

A class that cannot be instantiated directly and is meant to be subclassed. It's created using the `abc` module and can define abstract methods that subclasses *must* implement.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

### 17. What is the difference between a class variable and an instance variable?

- A **class variable** is shared across all instances of a class (defined at class level).
- An **instance variable** is unique to each object (defined inside `__init__` with `self`).

```python
class Dog:
    species = "Canine"        # class variable, shared

    def __init__(self, name):
        self.name = name       # instance variable, unique per object
```

### 18. What is multiple inheritance in Python?

When a class inherits from more than one parent class, combining their attributes and methods.

```python
class Flyer:
    def fly(self):
        print("Flying")

class Swimmer:
    def swim(self):
        print("Swimming")

class Duck(Flyer, Swimmer):   # multiple inheritance

Duck().fly(); Duck().swim()
```

### 19. Explain the purpose of `__str__` and `__repr__` methods in Python.

- `__str__` returns a human-readable, informal string representation (used by `print()` and `str()`).
- `__repr__` returns an unambiguous, developer-facing representation (used in the REPL/debugger, ideally something that could recreate the object).

```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y
    def __str__(self):
        return f"Point({self.x}, {self.y})"
    def __repr__(self):
        return f"Point(x={self.x!r}, y={self.y!r})"
```

### 20. What is the significance of the `super()` function in Python?

`super()` gives access to methods of a parent class from within a child class — commonly used to call the parent's `__init__` or extend (rather than fully override) a method.

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # calls Animal's __init__
        self.breed = breed
```

### 21. What is the significance of the `__del__` method in Python?

`__del__` is the destructor — called when an object is about to be garbage collected, typically used to release external resources (files, connections). Its exact timing isn't guaranteed, so it's used sparingly.

```python
class Resource:
    def __del__(self):
        print("Resource cleaned up")
```

### 22. What is the difference between @staticmethod and @classmethod in Python?

- `@staticmethod`: no implicit first argument; can't access class or instance state.
- `@classmethod`: receives `cls` as the first argument; can access/modify class-level state and is often used for alternative constructors.

```python
class Person:
    count = 0

    @classmethod
    def increment(cls):
        cls.count += 1

    @staticmethod
    def is_adult(age):
        return age >= 18
```

### 23. How does polymorphism work in Python with inheritance?

When subclasses override a parent's method, calling that method on a parent-typed reference automatically executes the subclass's version at runtime (dynamic dispatch).

```python
class Bird:
    def fly(self):
        print("Flies generically")

class Penguin(Bird):
    def fly(self):
        print("Can't fly!")

for b in [Bird(), Penguin()]:
    b.fly()   # different output per actual object type
```

### 24. What is method chaining in Python OOP?

A technique where each method returns `self` (the object itself), allowing multiple method calls to be *chained* together in a single statement.

```python
class Builder:
    def __init__(self):
        self.parts = []
    def add(self, part):
        self.parts.append(part)
        return self             # enables chaining

b = Builder().add("wheels").add("engine").add("seats")
print(b.parts)
```

### 25. What is the purpose of the `__call__` method in Python?

`__call__` lets an instance of a class be **called like a function**, i.e. `obj()`.

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    def __call__(self, x):
        return x * self.factor

double = Multiplier(2)
print(double(5))   # 10, obj called like a function
```

### 2. Abstract class `Shape` with `area()`; `Circle` and `Rectangle` implement it

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14159 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

c = Circle(5)
r = Rectangle(4, 6)
print("Circle area:", c.area())
print("Rectangle area:", r.area())
````

### 3. Multi-level inheritance: `Vehicle` → `Car` → `ElectricCar`

````python
class Vehicle:
    def __init__(self, type_):
        self.type = type_

class Car(Vehicle):
    def __init__(self, type_, brand):
        super().__init__(type_)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type_, brand, battery):
        super().__init__(type_, brand)
        self.battery = battery

ec = ElectricCar("Electric", "Tesla", "100 kWh")
print(ec.type, ec.brand, ec.battery)
````

### 4. Polymorphism: base class `Bird`, derived `Sparrow` and `Penguin` override `fly()`

````python
class Bird:
    def fly(self):
        print("Birds can generally fly.")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky.")

class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly, they swim instead.")

for bird in [Sparrow(), Penguin()]:
    bird.fly()
````

### 5. Encapsulation: `BankAccount` with private `balance`, deposit/withdraw/check methods

````python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance   # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance.")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")

    def check_balance(self):
        return self.__balance

acc = BankAccount(100)
acc.deposit(50)
acc.withdraw(30)
print("Balance:", acc.check_balance())
````

### 10. Operator overloading: `Vector` class overriding `__add__`

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 1)
print(v1 + v2)
````

### 12. `Student` class with `name`, `grades`, and `average_grade()`

````python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades) if self.grades else 0

s = Student("Sahil", [85, 90, 78, 92])
print(f"{s.name}'s average grade: {s.average_grade():.2f}")
````

### 13. `Rectangle` class with `set_dimensions()` and `area()`

### 14. `Employee` with `calculate_salary()`; derived `Manager` adds a bonus

````python
class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

emp = Employee(160, 20)
mgr = Manager(160, 30, 500)
print("Employee salary:", emp.calculate_salary())
print("Manager salary:", mgr.calculate_salary())
````

### 16. `Animal` with abstract method `sound()`; derived `Cow` and `Sheep`

````python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"

for animal in [Cow(), Sheep()]:
    print(animal.sound())
````
